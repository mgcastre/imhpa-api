# Ejemplo #1

En este ejemplo demostramos cómo descargar datos de un sensor en específico, de un sensor en específico.

```python
# Load libraries
import imhpa

# Initialize the client
client = imhpa.ImhpaClient()

# Get a list of all sensors
client.get_sensors()
```

```python
# Get all rain gages (pluviometros)
rain_gages = client.get_stations(sensor="LLUVIA")

# Write rain gages to local csv file
rain_gages.to_csv("../local/rain_gages.csv", index=False)
```

```python
# Get all rain gages in the San Pablo (No. 118) watershed
mask = rain_gages['nombre'].str.contains("118")
rain_gages.loc[mask, :]
```

```python
# Obtain data from station "EL MARIA"
data = client.get_data(sensor='LLUVIA', station_id='118005')

# Save data to a feather file
data.to_feather("../local/rain_data.ftr")
```
